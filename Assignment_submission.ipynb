```python
#writing a class containg variout functions such as Random , visualise, and pie
class random_gen:
  def __init__(self,n):
    #Xo = 5
    self.n = n

  def Random(self,decimal=None):
    import os
    import numpy as np
    #Xo = 5
    c = 3
    a = 3
    m = 50
    size = 3
    randomNums = [0] * (self.n)
    #randomNums[0] = Xo
    if decimal is None :
      decimal = 1
      l = 1
    else :
      l = 100000

    for i in range(1, self.n):
      result = os.urandom(size)
      # Follow the linear congruential method
      randomNums[i] = round((int.from_bytes(result,'little')/m + 1)/l , decimal)
    return randomNums

  def visualise(self):
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    import math
    X , Y = np.array(random_gen(self.n).Random(decimal=3)),np.array(random_gen(self.n).Random(decimal=3))

    plt.scatter(X,Y,color='b')
    plt.hlines(y=1,xmin=0,xmax=1)
    plt.vlines(x=1,ymin=0,ymax=1)
    plt.plot([1],[1], marker="o", markersize=7, markeredgecolor="red", markerfacecolor="black")

    plt.ylim((0,2))
    plt.xlim((0,2))
    plt.tight_layout()
    return X , Y 

  def pie(self):
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    import math
    circle = []
    X , Y = random_gen(self.n).visualise()
    cnt=0
    for i in list(zip(X,Y)):
      x , y = (i[0] - 1) , (i[1] - 1)
      dist = math.sqrt(x**2 + y**2)
      if dist <=1:
        circle.append(i)
      elif (dist <= math.sqrt(2)) & (dist > 1):
        cnt+=1
    a , b = list(zip(*circle))
    plt.scatter(a,b,color='r')
    plt.show()
    Pi = 4 * len(circle)/(len(circle)+cnt)  # 4 * (number of points inside the circle) / (Number of points points in the square)
    print(f"Estimated value of Pie is : {Pi}")
    return Pi
```
